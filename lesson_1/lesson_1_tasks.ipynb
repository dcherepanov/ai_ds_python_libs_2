1. Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.

1. При разработке моделей, которые позволяли бы реализовывать механизмы по замедлению реакции распада на ядерных электростанциях в случае возникновения каких-либо ошибок в управлении или неполадок, лучше максимизировать Recall, чтобы не допустить неконтролируемого распада, и лишний раз остановить работу станции.
2. А если лучше выбрать персонал по их квалификации для управления подобным объектом, то лучше использовать Precision.
3. При разработке моделей, направленных на поиск неисправностей в каких-либо механизмах, лучше максимизировать Recall, чтобы не упустить поломку.
4. При разработке моделей для поиска, скажем, дефектных деталей при их производстве, лучше максимизировать Precision, чтобы не допустить пропуска брака.

2. Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

F-мера позволяет в комплексе оценивать как Precision, так и Recall. Если взять просто среднее значение от Precision и Recall, то возможны варианты, когда при их изменении мы не увидим изменения F-меры. Кроме того, можно подбором параметра $\beta$ смещать влияние на F-меру в сторону либо Precision, либо Recall.
