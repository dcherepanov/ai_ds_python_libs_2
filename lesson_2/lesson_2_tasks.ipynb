1. В чём различие между зависимыми и независимыми выборками?

Если каждый случай одной выборки соостветствует каждому случаю из другой, то эти выборки зависимы. Если взаимосвязи между выборками нет, то они независимы. Например, мужья и жены - зависимые выборки, а мужчины женщины - независимые.

2. Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

Параметрические статистические критерии требуют знания закона распределения, а непараметрические не требуют. Например, при применении Хи-квадрат Пирсона, работают с частотами наблюдаемых параметров, а F-критерий требует знания дисперсии, а значит и закона распределения.
